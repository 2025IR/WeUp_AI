```python
#tool_calling training
import json, re, ast, importlib
from typing import List, Dict, Any
from datasets import load_dataset, Features, Value

# ────────────────────────── 토큰 상수
BOS  = "<|begin_of_text|>"
HDR  = "<|start_header_id|>"
ENDH = "<|end_header_id|>"
EOT  = "<|eot_id|>"

def _hdr(role: str, content: str) -> str:
    return f"{HDR}{role}{ENDH}\n{content}{EOT}\n"

# ────────────────────────── robust JSON 파서
_json5_spec = importlib.util.find_spec("json5")
if _json5_spec:
    import json5                                   # type: ignore

TRAIL_COMMA = re.compile(r",\s*([}\]])")

def _strip_commas(txt: str) -> str:
    return TRAIL_COMMA.sub(r"\1", txt)

def safe_json_parse(raw: Any) -> Dict[str, Any] | None:
    """다양한 비표준 schema 문자열 → dict (실패 시 None)."""
    if isinstance(raw, dict):
        return raw
    s = str(raw).strip()
    for parser in (
        lambda t: json.loads(t),
        (lambda t: json5.loads(t)) if _json5_spec else None,
        lambda t: json.loads(_strip_commas(t)),
        lambda t: ast.literal_eval(_strip_commas(t)),
    ):
        if parser is None:
            continue
        try:
            return parser(s)
        except Exception:
            pass
    return None
system_text ="""당신은 친절하고 신뢰할 수 있는 프로젝트 비서입니다.
                    사용자의 지시를 정확히 이해하고, 그 의도에 따라 적절한 행동을 능동적으로 수행해야 합니다.

                    다음 사항을 반드시 지켜주세요:
                    1. 말투는 **항상 공손하고 따뜻하게**, 사용자와 자연스럽고 부드러운 대화를 유지합니다.
                    2. 사용자의 질문이 프로젝트 진행, 개발, 일정, 협업 등과 관련된 경우, **효율적인 실행을 돕는 방식으로 안내**합니다.
                    3. 사용자의 질문 또는 요청이 명확하지 않더라도 **의도를 추론하여** 다음 중 적절한 행동을 취합니다:
                    - 직접 응답
                    - 함수 또는 API 호출
                    - 일상적인 대화 대응
                    4. 근거 없는 정보를 생성하지 않습니다. **정확한 정보에 기반해 답변하며**, 확실하지 않은 경우 “정확한 정보를 찾을 수 없습니다”라고 정중히 안내합니다.
                    5. 질문이 외부 지식에 기반해야 하는 경우, 스스로 정보를 생성하는 것이 아니라, **검색이나 문서 기반 질의 응답 시스템을 통해 확인된 정보만 사용**합니다.

                    사용자의 목적은 항상 “더 나은 프로젝트 진행”이므로, 그 목적에 부합하도록 지원하는 것이 당신의 가장 중요한 역할입니다."""


# ────────────────────────── prompt 빌더
# def build_prompt(func_desc: Dict[str, Any],
#                  user_msg: str,
#                  context: str,
#                  chat_hist: str) -> str:
#     system_txt = (
#         "You are an assistant that can call the following function.\n"
#         + json.dumps(func_desc, ensure_ascii=False)
#     )
#     return (
#         f"{BOS}{_hdr('system', system_txt)}"
#         f"{_hdr('context', context)}"
#         f"{_hdr('chat_history', chat_hist)}"
#         f"{_hdr('user', user_msg)}"
#         f"{HDR}assistant{ENDH}\n"
#     )
def build_prompt(
    user_system_text: str,               # ← 사용자가 지정한 시스템 지침
    func_desc: Dict[str, Any],           # ← 함수 스키마(dict)
    user_msg: str,
    chat_hist: str = "",
) -> str:
    """
    system  : 사용자의 톤·역할 규칙 (고정)
    context : 이번 세션에서 호출 가능한 함수 JSON
    """
    system_block  = _hdr("system", user_system_text.strip())
    context_block = _hdr("context", json.dumps(func_desc, ensure_ascii=False))
    history_block = _hdr("chat_history", chat_hist)
    user_block    = _hdr("user", user_msg)

    return (
        f"{BOS}{system_block}"
        f"{context_block}"
        f"{history_block}"
        f"{user_block}"
        f"{HDR}assistant{ENDH}\n"
    )
# ────────────────────────── 대화 파서
ROLE_TAG = {
    "user":              "<usr>",
    "assistant":         "<bot>",
    "function-call":     "<function-call>",
    "function-response": "<function-response>",
}

def parse_dialog(conv: List[Dict[str, str]]):
    try:
        last_a = max(i for i,m in enumerate(conv) if m["role"]=="assistant")
        last_u = max(i for i,m in enumerate(conv[:last_a]) if m["role"]=="user")
    except ValueError:
        return None
    hist   = conv[:last_u]
    user   = conv[last_u]["content"].strip()
    answer = conv[last_a]["content"].strip()
    hist_str = "\n".join(
        f"{ROLE_TAG.get(m['role'], '<unk>')} {m['content'].strip()}" for m in hist
    )
    return hist_str, user, answer

# ────────────────────────── 레코드 → prompt / target
def make_prompt_target(example: Dict[str, Any]) -> Dict[str, str | None]:
    hist_parse = parse_dialog(example["conversations_ko"])
    func_desc  = safe_json_parse(example["function_description"])
    if hist_parse is None or func_desc is None:
        return {"prompt": None, "completion": None}

    hist, user_msg, answer = hist_parse
    prompt = build_prompt(
    user_system_text = system_text,   # ① system 지침
    func_desc        = func_desc,     # ② 함수 스키마
    user_msg         = user_msg,      # ③ 현재 사용자 발화
    chat_hist        = hist           # ④ 이전 대화
)
    return {"prompt": prompt, "completion": answer + EOT}

# ────────────────────────── 데이터셋 변환
raw_train = load_dataset("heegyu/glaive-function-calling-v2-ko", split="train")

# 새 스키마: prompt / target 두 컬럼만 남김
new_feats = Features({"prompt": Value("string"), "completion": Value("string")})

tmp = raw_train.map(
    make_prompt_target,
    num_proc=4,                        # 디버깅 시 1 → 문제 없으면 4~8
    desc="build prompts",
    features=new_feats,                # 두 컬럼 고정
    remove_columns=raw_train.column_names  # ★ 원본 컬럼 전부 제거 ★
)

processed = tmp.filter(lambda x: x["prompt"] is not None, desc="drop invalid")

print(f"✅ 변환 완료: {len(processed):,} samples")
print(processed[0]["prompt"][:6000], "...\n▶ target:", processed[0]["completion"][:200], "…")
```

```python
from typing import Tuple, Union
import re
import json
from peft import PeftModel
from transformers import AutoModelForCausalLM

# --- Configuration Constants ---
SYSTEM_TEXT = r"""
당신은 친절하고 신뢰할 수 있는 프로젝트 비서입니다.
사용자의 지시를 정확히 이해하고, 그 의도에 따라 적절한 행동을 능동적으로 수행해야 합니다.
사용자의 목적은 항상 “더 나은 프로젝트 진행”이므로, 그 목적에 부합하도록 지원하는 것이 당신의 가장 중요한 역할입니다.

다음 사항을 반드시 지켜주세요:
1. 말투는 항상 공손하고 따뜻하게, 사용자와 자연스럽고 부드러운 대화를 유지합니다.
2. 사용자의 질문이 프로젝트 진행, 개발, 일정, 협업 등과 관련된 경우, 효율적인 실행을 돕는 방식으로 안내합니다.
3. 사용자의 질문 또는 요청이 명확하지 않더라도 의도를 추론하여 다음 중 적절한 행동을 취합니다:
   - 직접 응답
   - 함수 또는 API 호출
   - 일상적인 대화 대응
4. 근거 없는 정보를 생성하지 않습니다. 정확한 정보에 기반해 답변하며, 확실하지 않은 경우 “정확한 정보를 찾을 수 없습니다”라고 정중히 안내합니다.
5. 질문이 외부 지식에 기반해야 하는 경우, 스스로 정보를 생성하는 것이 아니라, 검색이나 문서 기반 질의 응답 시스템을 통해 확인된 정보만 사용합니다.
6. 질문이 여러 개 있더라도 한 번에 하나만 응답합니다.
"""

AGENT_TEXT = r"""
※ 출력 예시:
Thought: 회의 일정을 등록해야 합니다.
Action: record_meeting
Action Input: {"project_id": 12, "title": "팀 미팅", "date": "2023-05-24", "time": "10:00"}
Observation: 회의가 등록되었습니다.
Final Answer: 2023년 5월 24일 10시에 팀 미팅이 성공적으로 등록되었습니다.
자세한 내용은 다음 프롬프트를 기다립니다.
"""

# FUNCTION_SCHEMAS should be defined or imported from your application context
FUNCTION_SCHEMAS = {...}

# Keywords triggering agent mode
AGENT_KEYWORDS = ["회의", "등록", "일정", "잡아줘", "미팅", "변경"]

class PromptBuilder:
    BOS = "<|begin_of_text|>"
    HDR = "<|start_header_id|>"
    ENDH = "<|end_header_id|>"
    EOT = "<|eot_id|>"

    def __init__(self, system_text: str, agent_text: str, context_schema: str):
        self.system_text = system_text
        self.agent_text = agent_text
        self.context_schema = context_schema

    def build(self, user_msg: str, use_agent: bool, chat_history: str = "") -> str:
        system_block = self.system_text + (self.agent_text if use_agent else "")
        parts = [
            f"{self.BOS}{self.HDR}system{self.ENDH}\n{system_block}{self.EOT}",
        ]
        if use_agent:
            parts.append(f"{self.HDR}context{self.ENDH}\n{self.context_schema}{self.EOT}")
        parts.append(f"{self.HDR}chat_history{self.ENDH}\n{chat_history}{self.EOT}")
        parts.append(f"{self.HDR}user{self.ENDH}\n{user_msg}{self.EOT}")
        parts.append(f"{self.HDR}assistant{self.ENDH}\n")
        return "\n".join(parts)

class ActionParser:
    @staticmethod
    def parse(text: str) -> Tuple[Union[str, None], Union[dict, str, None], Union[str, None]]:
        # Extract Action
        action_m = re.search(r'Action:\s*(\w+)', text)
        action = action_m.group(1) if action_m else None
        # Extract Action Input
        input_m = re.search(r'Action Input:\s*(\{.*?\})(?=\s*(Observation:|Final Answer:))', text, re.DOTALL)
        if input_m:
            raw = input_m.group(1)
            try:
                action_input = json.loads(raw)
            except json.JSONDecodeError:
                action_input = raw
        else:
            action_input = None
        # Extract Final Answer
        fa_m = re.search(r'Final Answer:\s*(.+)', text)
        final_answer = fa_m.group(1).strip() if fa_m else None
        return action, action_input, final_answer

class ModelManager:
    def __init__(self, base_model_name: str, adapter_paths: dict, device_map="auto", quantization_config=None):
        self.base = AutoModelForCausalLM.from_pretrained(
            base_model_name,
            device_map=device_map,
            quantization_config=quantization_config
        )
        self.adapter_paths = adapter_paths

    def get_adapter(self, name: str, trainable: bool = False):
        return PeftModel.from_pretrained(
            self.base,
            self.adapter_paths[name],
            adapter_name=name,
            is_trainable=trainable
        )

class ProjectAssistant:
    def __init__(self, model_manager: ModelManager, prompt_builder: PromptBuilder):
        self.mm = model_manager
        self.pb = prompt_builder

    def handle_input(self, text_input: str):
        use_agent = any(k in text_input for k in AGENT_KEYWORDS)
        adapter = "agent" if use_agent else "multi_turn"
        model = self.mm.get_adapter(adapter)
        prompt = self.pb.build(text_input, use_agent)
        response = self._generate(prompt, model)

        if use_agent:
            action, params, final_resp = ActionParser.parse(response)
            self._execute(action, params)
            print(response)
        else:
            print(response)

    @staticmethod
    def _generate(prompt: str, model) -> str:
        inputs = tok(prompt, return_tensors="pt").to(model.device)
        out_ids = model.generate(
            **inputs,
            max_new_tokens=256,
            do_sample=True,
            temperature=0.8,
            top_p=0.8,
            eos_token_id=tok.eos_token_id
        )
        return tok.decode(out_ids[0], skip_special_tokens=True)

    @staticmethod
    def _execute(action: str, params: Union[dict, str, None]):
        if action == "regist_todo" and isinstance(params, dict):
            print(regist_todo(**params))
        elif action == "change_role" and isinstance(params, dict):
            print(change_role(**params))

# --- Entry Point ---
def main():
    user_input = input("요청 사항을 입력해주세요\n")
    # Initialize components
    prompt_builder = PromptBuilder(SYSTEM_TEXT, AGENT_TEXT, FUNCTION_SCHEMAS)
    adapter_paths = {
        "agent": "/content/drive/MyDrive/tool_call_epoch5/agent",
        "multi_turn": "/content/drive/MyDrive/bllossom-multiturn-epoch5"
    }
    model_manager = ModelManager(
        "MLP-KTLim/llama-3-Korean-Bllossom-8B",
        adapter_paths,
        quantization_config=bnb_config
    )
    assistant = ProjectAssistant(model_manager, prompt_builder)
    assistant.handle_input(user_input)

if __name__ == "__main__":
    main()
```

```python
#multi_turn training

torch.backends.cuda.enable_flash_sdp(False)
torch.backends.cuda.enable_mem_efficient_sdp(False)
torch.backends.cuda.enable_math_sdp(True)

bnb_config = BitsAndBytesConfig(
    load_in_4bit=True,
    bnb_4bit_quant_type="nf4",
    bnb_4bit_compute_dtype=torch.bfloat16
)
lora_config = LoraConfig(
    r=16,
    lora_alpha=32,
    lora_dropout=0.05,
    bias="none",
    target_modules=["q_proj", "o_proj", "k_proj", "v_proj",
                    "gate_proj", "up_proj", "down_proj"],
    task_type=TaskType.CAUSAL_LM
)
def build_prompt(system_text, user_instruction, context="", chat_history=""):
    """
    LLaMA3 시스템/유저/어시스턴트 구조로 한 줄 프롬프트를 만듦.
    assistant_output이 있으면 '정답'까지 포함한 완성된 시퀀스.
    """
    prompt = (
        f"<|begin_of_text|><|start_header_id|>system<|end_header_id|> {system_text}<|eot_id|>\n"
        f"<|start_header_id|>context<|end_header_id|>\n{context}<|eot_id|>\n"
        f"<|start_header_id|>chat_history<|end_header_id|>\n{chat_history}<|eot_id|>\n"
        f"<|start_header_id|>user<|end_header_id|>\n{user_instruction}<|eot_id|>\n"
        f"<|start_header_id|>assistant<|end_header_id|>\n"
    )
    return prompt

def parse_dialog_data(raw_text):
    """
    0) A: 또는 A->, B: 또는 B-> 형식을 <usr>/<bot> 태그로 변환
    1) <usr>/<bot> 태그로 분리
    2) 잘못된 페어면 None 반환 → 스킵
    """
    # 0. A/B 포맷 치환 (멀티라인 모드)
    raw_text = re.sub(r'(?m)^[ \t]*A\s*(?:->|:)\s*', '<usr> ', raw_text)
    raw_text = re.sub(r'(?m)^[ \t]*B\s*(?:->|:)\s*', '<bot> ', raw_text)

    # 1. <usr> 태그 기준으로 context / chat 분리
    if "<usr>" not in raw_text or "<bot>" not in raw_text:
        return None
    try:
        context, chat_part = raw_text.split("<usr>", 1)
    except ValueError:
        return None
    context = context.replace("<sys>", "").strip()

    # 2. chat_part 에서 <usr>/<bot> 페어 파싱
    turns = chat_part.split("<usr>")
    dialog = []
    for turn in turns:
        turn = turn.strip()
        if not turn:
            continue
        if "<bot>" in turn:
            user_part, bot_part = turn.split("<bot>", 1)
            dialog.append(("<usr>", user_part.strip()))
            dialog.append(("<bot>", bot_part.strip()))
        else:
            # <bot> 없는 user-only 턴은 스킵
            return None

    # 3. 마지막 턴이 제대로 페어를 이루는지 확인
    if len(dialog) < 2 or dialog[-2][0] != "<usr>" or dialog[-1][0] != "<bot>":
        return None

    # 4. 결과 분리
    last_user = dialog[-2][1]
    last_bot  = dialog[-1][1]
    chat_history = dialog[:-2]
    chat_history_str = "\n".join(f"{role} {text}" for role, text in chat_history)

    return {
        "context":         context,
        "chat_history":    chat_history_str,
        "user_instruction": last_user,
        "assistant_output": last_bot
    }

def safe_multiturn_formatting_func(example):
    try:
        return multiturn_formatting_func(example)
    except Exception as e:
        print(f"❌ Skipping example due to error: {e}")
        return None  # 이 경우는 `load_dataset(...).map(..., remove_columns=None)`에서 자동 필터링됨

def multiturn_formatting_func(example):
    """
    SFTTrainer가 사용하는 포맷 함수.
    - 출력값은 문자열이 아닌 tokenizer 입력에 맞춘 dict
    - train_on_inputs=False의 효과를 반영: 프롬프트 부분을 loss 계산에서 제외
    """
    output_texts = []
    # System 메시지 (사용자 정의 가능)
    system_text ="""당신은 친절하고 신뢰할 수 있는 프로젝트 비서입니다.
                    사용자의 지시를 정확히 이해하고, 그 의도에 따라 적절한 행동을 능동적으로 수행해야 합니다.

                    다음 사항을 반드시 지켜주세요:
                    1. 말투는 **항상 공손하고 따뜻하게**, 사용자와 자연스럽고 부드러운 대화를 유지합니다.
                    2. 사용자의 질문이 프로젝트 진행, 개발, 일정, 협업 등과 관련된 경우, **효율적인 실행을 돕는 방식으로 안내**합니다.
                    3. 사용자의 질문 또는 요청이 명확하지 않더라도 **의도를 추론하여** 다음 중 적절한 행동을 취합니다:
                    - 직접 응답
                    - 함수 또는 API 호출
                    - 일상적인 대화 대응
                    4. 근거 없는 정보를 생성하지 않습니다. **정확한 정보에 기반해 답변하며**, 확실하지 않은 경우 “정확한 정보를 찾을 수 없습니다”라고 정중히 안내합니다.
                    5. 질문이 외부 지식에 기반해야 하는 경우, 스스로 정보를 생성하는 것이 아니라, **검색이나 문서 기반 질의 응답 시스템을 통해 확인된 정보만 사용**합니다.

                    사용자의 목적은 항상 “더 나은 프로젝트 진행”이므로, 그 목적에 부합하도록 지원하는 것이 당신의 가장 중요한 역할입니다."""

    # 사용자 질문 구성
    datas = parse_dialog_data(example['text'])
    if datas == None:
        return None
    context = datas['context']
    chat_history = datas['chat_history']
    user_instruction = datas['user_instruction']
    assistant_output = datas['assistant_output']
    prompt = build_prompt(system_text, user_instruction, context, chat_history)
    full_text = prompt +assistant_output + "<|eot_id|>"
    tokenized = tokenizer(
        full_text,
        padding="max_length",
        max_length=1024,
        truncation=True,
        return_tensors="pt",
    )
    input_ids = tokenized["input_ids"][0].tolist()
    labels = input_ids[:]

    prompt_ids = tokenizer(prompt,add_special_tokens=False).input_ids
    prompt_len = len(prompt_ids)

    for i in range(min(prompt_len, len(labels))):
        labels[i] = -100

    tokenized["labels"] = torch.tensor([labels])

    return {
        "input_ids": tokenized["input_ids"][0],
        "attention_mask": tokenized["attention_mask"][0],
        "labels": tokenized["labels"][0],
    }

print("Model loading...")
base_model = AutoModelForCausalLM.from_pretrained(
    "MLP-KTLim/llama-3-Korean-Bllossom-8B",
    device_map="auto",    # 단일 GPU 사용 예시
    quantization_config=bnb_config
)

model = get_peft_model(base_model, lora_config)
model.print_trainable_parameters()

tokenizer = AutoTokenizer.from_pretrained("MLP-KTLim/llama-3-Korean-Bllossom-8B")
tokenizer.add_special_tokens({
    "additional_special_tokens": [
        "<|start_header_id|>", "<|end_header_id|>"
    ]
})
tokenizer.padding_side = 'right'

training_param={
    "epoch": 5,
    "batch_size": 8,
    "learning_rate": 5e-6,
    "weight_decay": 0.01,
    "warmup_steps": 300,
}

folder_name = f"bllossom_model/bllossom-multiturn-epoch2-chain_prompt"
wandb.init(
    project="bllossom",
    name= "only multiturn",
)

multiturn_dataset = load_dataset("heegyu/korquad-chat-v1")
multiturn_train_data = multiturn_dataset['train']
print(type(multiturn_train_data))

processed_list = multiturn_train_data.map(multiturn_formatting_func, remove_columns=multiturn_train_data.column_names)
processed_list = [ex for ex in processed_list if ex is not None]
dataset = Dataset.from_list(processed_list)
train_val_dataset, test_dataset = dataset.train_test_split(test_size=0.1, seed=42).values()
train_dataset, val_dataset = train_val_dataset.train_test_split(test_size=0.1111, seed=42).values()



multiturn_processed_data = multiturn_train_data.map(
    safe_multiturn_formatting_func,
    batched=False,
    remove_columns=multiturn_train_data.column_names
)

mutiturn_trainer = SFTTrainer(
    model=model,
    formatting_func = multiturn_formatting_func,
    train_dataset=train_dataset,
    eval_dataset=val_dataset,
    args=TrainingArguments(
        output_dir="outputs",
        num_train_epochs=training_param['epoch'],
        per_device_train_batch_size=1,
        gradient_accumulation_steps=training_param['batch_size'],
        optim="adamw_torch",
        lr_scheduler_type="cosine",
        weight_decay=training_param['weight_decay'],
        warmup_steps=training_param['warmup_steps'],
        learning_rate=training_param['learning_rate'],
        fp16=True,
        bf16=False,
        max_grad_norm=0.5,
        evaluation_strategy="epoch",
        eval_steps=100,
        logging_steps=10,
        push_to_hub=False,
        save_steps=500,
        save_total_limit=2,
        report_to=["wandb"],
    ),
)

mutiturn_trainer.train()
mutiturn_trainer.model.save_pretrained(folder_name)

val_metrics = mutiturn_trainer.evaluate(eval_dataset=val_dataset)
test_metrics = mutiturn_trainer.evaluate(eval_dataset=test_dataset)

wandb.log({
    "eval_loss": val_metrics["eval_loss"],
    "test_loss": test_metrics["eval_loss"]
})
```

```python
def route(text: str) -> str:
    """
    아주 단순한 휴리스틱:
    - project_id, 역할, 회의 같은 키워드가 있고 JSON 패턴이 예상되면 'agent'
    - 아니면 'chat'
    필요 시 LLM 기반 분류 체인으로 교체 가능
    """
    keywords = ["project_id", "역할", "회의", "일정", "등록", "추가"]
    if any(k in text for k in keywords):
        return "agent"
    return "chat"
```

```python
class ActionParser(AgentOutputParser):
    def parse(self, text: str):
        pattern = (r"Action:\s*(\w+).*?"
                   r"Action Input:\s*(\{.*?\}).*?"
                   r"Final Answer:(.*)")
        match = re.search(pattern, text, re.S)
        if not match:
            return AgentFinish(return_values={"output": text},
                               log=text)
        action, js, answer = match.groups()
        try:
            inp = json.loads(js)
        except json.JSONDecodeError:
            return AgentFinish(return_values={"output": text},
                               log=text)
        return AgentAction(tool=action,
                           tool_input=inp,
                           log=text)

parser = ActionParser()
```

```python
# -----------------------------
# 2) 데이터셋 로드
#    KoAlpaca-v1.1a는 {"instruction", "input", "output"} 구조
# -----------------------------
dataset = load_dataset('beomi/KoAlpaca-v1.1a')
train_data = dataset['train']
print(type(train_data))
# -----------------------------
# 3) LLaMA3 스타일 프롬프트 생성 함수
#    -> 단일 예제(Instruction/Output) 기준으로 문자열 1개를 반환
#    -> TRL SFTTrainer는 이 문자열을 토크나이징 + 부분 마스킹 수행
# -----------------------------
def build_prompt(system_text, user_instruction, context="", chat_history=""):
    """
    LLaMA3 시스템/유저/어시스턴트 구조로 한 줄 프롬프트를 만듦.
    assistant_output이 있으면 '정답'까지 포함한 완성된 시퀀스.
    """
    prompt = (
        f"<|begin_of_text|><|start_header_id|>system<|end_header_id|> {system_text}<|eot_id|>\n"
        f"<|start_header_id|>context<|end_header_id|>\n{context}<|eot_id|>\n"
        f"<|start_header_id|>chat_history<|end_header_id|>\n{chat_history}<|eot_id|>\n"
        f"<|start_header_id|>user<|end_header_id|>\n{user_instruction}<|eot_id|>\n"
        f"<|start_header_id|>assistant<|end_header_id|>\n"
    )
    return prompt


def formatting_func(example):
    """
    SFTTrainer가 사용하는 포맷 함수.
    - 출력값은 문자열이 아닌 tokenizer 입력에 맞춘 dict
    - train_on_inputs=False의 효과를 반영: 프롬프트 부분을 loss 계산에서 제외
    """
    output_texts = []
    # System 메시지 (사용자 정의 가능)
    system_text ="""당신은 친절하고 신뢰할 수 있는 프로젝트 비서입니다.
                    사용자의 지시를 정확히 이해하고, 그 의도에 따라 적절한 행동을 능동적으로 수행해야 합니다.

                    다음 사항을 반드시 지켜주세요:
                    1. 말투는 **항상 공손하고 따뜻하게**, 사용자와 자연스럽고 부드러운 대화를 유지합니다.
                    2. 사용자의 질문이 프로젝트 진행, 개발, 일정, 협업 등과 관련된 경우, **효율적인 실행을 돕는 방식으로 안내**합니다.
                    3. 사용자의 질문 또는 요청이 명확하지 않더라도 **의도를 추론하여** 다음 중 적절한 행동을 취합니다:
                    - 직접 응답
                    - 함수 또는 API 호출
                    - 일상적인 대화 대응
                    4. 근거 없는 정보를 생성하지 않습니다. **정확한 정보에 기반해 답변하며**, 확실하지 않은 경우 “정확한 정보를 찾을 수 없습니다”라고 정중히 안내합니다.
                    5. 질문이 외부 지식에 기반해야 하는 경우, 스스로 정보를 생성하는 것이 아니라, **검색이나 문서 기반 질의 응답 시스템을 통해 확인된 정보만 사용**합니다.

                    사용자의 목적은 항상 “더 나은 프로젝트 진행”이므로, 그 목적에 부합하도록 지원하는 것이 당신의 가장 중요한 역할입니다."""

    # 사용자 질문 구성
    user_instruction = example["instruction"]
    assistant_output = example["output"]
    prompt = build_prompt(system_text, user_instruction)
    full_text = prompt +assistant_output + "<|eot_id|>"
    tokenized = tokenizer(
        full_text,
        padding="max_length",
        max_length=512,
        truncation=True,
        return_tensors="pt",  # 단일 샘플이면 batched=False로 map 시 return_tensors="pt" 써도 무방
    )
    input_ids = tokenized["input_ids"][0].tolist()
    labels = input_ids[:]

    prompt_ids = tokenizer(prompt,add_special_tokens=False).input_ids
    prompt_len = len(prompt_ids)

    for i in range(min(prompt_len, len(labels))):
        labels[i] = -100

    tokenized["labels"] = torch.tensor([labels])

    return {
        "input_ids": tokenized["input_ids"][0],
        "attention_mask": tokenized["attention_mask"][0],
        "labels": tokenized["labels"][0],
    }

# -----------------------------
# 4) 모델 로드 + LoRA 설정
# -----------------------------
bnb_config = BitsAndBytesConfig(
    load_in_4bit=True,
    bnb_4bit_quant_type="nf4",
    bnb_4bit_compute_dtype=torch.bfloat16
)

# ==> 토크나이저 선언
tokenizer = AutoTokenizer.from_pretrained("MLP-KTLim/llama-3-Korean-Bllossom-8B")
tokenizer.add_special_tokens({
    "additional_special_tokens": [
        "<|start_header_id|>", "<|end_header_id|>"
    ]
})
tokenizer.padding_side = 'right'

print("Model loading...")
model = AutoModelForCausalLM.from_pretrained(
    "MLP-KTLim/llama-3-Korean-Bllossom-8B",
    device_map="auto",    # 단일 GPU 사용 예시
    quantization_config=bnb_config
)
model.resize_token_embeddings(len(tokenizer))  # 필요시 모델 임베딩 확장
lora_config = LoraConfig(
    r=16,
    lora_alpha=32,
    lora_dropout=0.05,
    bias="none",
    target_modules=["q_proj", "o_proj", "k_proj", "v_proj",
                    "gate_proj", "up_proj", "down_proj"],
    task_type=TaskType.CAUSAL_LM
)
model = get_peft_model(model, lora_config)
model.print_trainable_parameters()

training_param={
    "epoch": 3,
    "batch_size": 8,
    "learning_rate": 3e-5,
    "weight_decay": 0.01,
    "warmup_steps": 90,
}

folder_name = f"drive/MyDrive/bllossom-lora-long_prompt_masking-{training_param['epoch']}-{training_param['batch_size']}-{training_param['learning_rate']}-0419"
wandb.init(
    project="bllossom",
    name= folder_name
)

processed_data = train_data.map(formatting_func, batched=False)
# -----------------------------
# 5) SFTTrainer 설정
#    - formatting_func=... : 위에서 정의한 함수로 프롬프트 구성
#    - train_on_inputs=False : 지시문(프롬프트)에 대해서는 -100 마스킹
#    - tokenizer=...        : 'formatting_func' 결과문자열을 토크나이징할 때 사용
# -----------------------------
trainer = SFTTrainer(
    model=model,
    formatting_func = formatting_func,
    train_dataset=processed_data,
    max_seq_length=512,           # 토크나이저 최대 길이
    tokenizer=tokenizer,          # 꼭 추가해줘야 SFTTrainer가 tokenizer를 활용
    args=TrainingArguments(
        output_dir="outputs",
        num_train_epochs=training_param['epoch'],
        max_steps=3000,
        per_device_train_batch_size=1,
        gradient_accumulation_steps=training_param['batch_size'],
        optim="adamw_torch",
        lr_scheduler_type="cosine",
        weight_decay=0.01,
        warmup_steps=90,
        learning_rate=training_param['learning_rate'],
        fp16=True,
        max_grad_norm=0.3,
        logging_steps=10,
        push_to_hub=False,
        save_steps=500,
        save_total_limit=2,
        report_to=["wandb"],
    ),
)

# -----------------------------
# 6) 학습 수행
# -----------------------------
trainer.train()

trainer.model.save_pretrained(folder_name)
```

```python
from peft import PeftModel

bnb_config = BitsAndBytesConfig(
    load_in_4bit=True,
    bnb_4bit_quant_type="nf4",
    bnb_4bit_compute_dtype=torch.bfloat16
)
lora_config = LoraConfig(
    r=16,
    lora_alpha=32,
    lora_dropout=0.05,
    bias="none",
    target_modules=["q_proj", "o_proj", "k_proj", "v_proj",
                    "gate_proj", "up_proj", "down_proj"],
    task_type=TaskType.CAUSAL_LM
)
def build_prompt(system_text, user_instruction, context="", chat_history=""):
    """
    LLaMA3 시스템/유저/어시스턴트 구조로 한 줄 프롬프트를 만듦.
    assistant_output이 있으면 '정답'까지 포함한 완성된 시퀀스.
    """
    prompt = (
        f"<|begin_of_text|><|start_header_id|>system<|end_header_id|> {system_text}<|eot_id|>\n"
        f"<|start_header_id|>context<|end_header_id|>\n{context}<|eot_id|>\n"
        f"<|start_header_id|>chat_history<|end_header_id|>\n{chat_history}<|eot_id|>\n"
        f"<|start_header_id|>user<|end_header_id|>\n{user_instruction}<|eot_id|>\n"
        f"<|start_header_id|>assistant<|end_header_id|>\n"
    )
    return prompt

def parse_dialog_data(raw_text):
    """
    0) A: 또는 A->, B: 또는 B-> 형식을 <usr>/<bot> 태그로 변환
    1) <usr>/<bot> 태그로 분리
    2) 잘못된 페어면 None 반환 → 스킵
    """
    # 0. A/B 포맷 치환 (멀티라인 모드)
    raw_text = re.sub(r'(?m)^[ \t]*A\s*(?:->|:)\s*', '<usr> ', raw_text)
    raw_text = re.sub(r'(?m)^[ \t]*B\s*(?:->|:)\s*', '<bot> ', raw_text)

    # 1. <usr> 태그 기준으로 context / chat 분리
    if "<usr>" not in raw_text or "<bot>" not in raw_text:
        return None
    try:
        context, chat_part = raw_text.split("<usr>", 1)
    except ValueError:
        return None
    context = context.replace("<sys>", "").strip()

    # 2. chat_part 에서 <usr>/<bot> 페어 파싱
    turns = chat_part.split("<usr>")
    dialog = []
    for turn in turns:
        turn = turn.strip()
        if not turn:
            continue
        if "<bot>" in turn:
            user_part, bot_part = turn.split("<bot>", 1)
            dialog.append(("<usr>", user_part.strip()))
            dialog.append(("<bot>", bot_part.strip()))
        else:
            # <bot> 없는 user-only 턴은 스킵
            return None

    # 3. 마지막 턴이 제대로 페어를 이루는지 확인
    if len(dialog) < 2 or dialog[-2][0] != "<usr>" or dialog[-1][0] != "<bot>":
        return None

    # 4. 결과 분리
    last_user = dialog[-2][1]
    last_bot  = dialog[-1][1]
    chat_history = dialog[:-2]
    chat_history_str = "\n".join(f"{role} {text}" for role, text in chat_history)

    return {
        "context":         context,
        "chat_history":    chat_history_str,
        "user_instruction": last_user,
        "assistant_output": last_bot
    }

def safe_multiturn_formatting_func(example):
    try:
        return multiturn_formatting_func(example)
    except Exception as e:
        print(f"❌ Skipping example due to error: {e}")
        return None  # 이 경우는 `load_dataset(...).map(..., remove_columns=None)`에서 자동 필터링됨

def multiturn_formatting_func(example):
    """
    SFTTrainer가 사용하는 포맷 함수.
    - 출력값은 문자열이 아닌 tokenizer 입력에 맞춘 dict
    - train_on_inputs=False의 효과를 반영: 프롬프트 부분을 loss 계산에서 제외
    """
    output_texts = []
    # System 메시지 (사용자 정의 가능)
    system_text ="""당신은 친절하고 신뢰할 수 있는 프로젝트 비서입니다.
                    사용자의 지시를 정확히 이해하고, 그 의도에 따라 적절한 행동을 능동적으로 수행해야 합니다.

                    다음 사항을 반드시 지켜주세요:
                    1. 말투는 **항상 공손하고 따뜻하게**, 사용자와 자연스럽고 부드러운 대화를 유지합니다.
                    2. 사용자의 질문이 프로젝트 진행, 개발, 일정, 협업 등과 관련된 경우, **효율적인 실행을 돕는 방식으로 안내**합니다.
                    3. 사용자의 질문 또는 요청이 명확하지 않더라도 **의도를 추론하여** 다음 중 적절한 행동을 취합니다:
                    - 직접 응답
                    - 함수 또는 API 호출
                    - 일상적인 대화 대응
                    4. 근거 없는 정보를 생성하지 않습니다. **정확한 정보에 기반해 답변하며**, 확실하지 않은 경우 “정확한 정보를 찾을 수 없습니다”라고 정중히 안내합니다.
                    5. 질문이 외부 지식에 기반해야 하는 경우, 스스로 정보를 생성하는 것이 아니라, **검색이나 문서 기반 질의 응답 시스템을 통해 확인된 정보만 사용**합니다.

                    사용자의 목적은 항상 “더 나은 프로젝트 진행”이므로, 그 목적에 부합하도록 지원하는 것이 당신의 가장 중요한 역할입니다."""

    # 사용자 질문 구성
    datas = parse_dialog_data(example['text'])
    if datas == None:
        return None
    context = datas['context']
    chat_history = datas['chat_history']
    user_instruction = datas['user_instruction']
    assistant_output = datas['assistant_output']
    prompt = build_prompt(system_text, user_instruction, context, chat_history)
    full_text = prompt +assistant_output + "<|eot_id|>"
    tokenized = tokenizer(
        full_text,
        padding="max_length",
        max_length=512,
        truncation=True,
        return_tensors="pt",
    )
    input_ids = tokenized["input_ids"][0].tolist()
    labels = input_ids[:]

    prompt_ids = tokenizer(prompt,add_special_tokens=False).input_ids
    prompt_len = len(prompt_ids)

    for i in range(min(prompt_len, len(labels))):
        labels[i] = -100

    tokenized["labels"] = torch.tensor([labels])

    return {
        "input_ids": tokenized["input_ids"][0],
        "attention_mask": tokenized["attention_mask"][0],
        "labels": tokenized["labels"][0],
    }

print("Model loading...")
base_model = AutoModelForCausalLM.from_pretrained(
    "MLP-KTLim/llama-3-Korean-Bllossom-8B",
    device_map={"":0},    # 단일 GPU 사용 예시
    quantization_config=bnb_config
)

model = get_peft_model(base_model, lora_config)
model.print_trainable_parameters()

tokenizer = AutoTokenizer.from_pretrained("MLP-KTLim/llama-3-Korean-Bllossom-8B")
tokenizer.add_special_tokens({
    "additional_special_tokens": [
        "<|start_header_id|>", "<|end_header_id|>"
    ]
})
tokenizer.padding_side = 'right'
# lora_model_path = "drive/MyDrive/bllossom-lora-long_prompt_masking-3-8-3e-05-0419"
# model = get_peft_model(base_model, lora_config)
# model.load_adapter(lora_model_path, adapter_name="default")
# model.set_adapter("default")
# model.print_trainable_parameters()


training_param={
    "epoch": 3,
    "batch_size": 8,
    "learning_rate": 5e-5,
    "weight_decay": 0.01,
    "warmup_steps": 150,
}

folder_name = f"drive/MyDrive/bllossom-lora-long_prompt_masking-{training_param['epoch']}-{training_param['batch_size']}-{training_param['learning_rate']}-multiturn"
wandb.init(
    project="bllossom",
    name= "only multiturn"
)

multiturn_dataset = load_dataset("heegyu/korquad-chat-v1")
multiturn_train_data = multiturn_dataset['train']
print(type(multiturn_train_data))

multiturn_processed_data = multiturn_train_data.map(
    safe_multiturn_formatting_func,
    batched=False,
    remove_columns=multiturn_train_data.column_names
)

mutiturn_trainer = SFTTrainer(
    model=model,
    formatting_func = multiturn_formatting_func,
    train_dataset=multiturn_processed_data,
    args=TrainingArguments(
        output_dir="outputs",
        num_train_epochs=training_param['epoch'],
        max_steps=3000,
        per_device_train_batch_size=1,
        gradient_accumulation_steps=training_param['batch_size'],
        optim="adamw_torch",
        lr_scheduler_type="cosine",
        weight_decay=0.01,
        warmup_steps=90,
        learning_rate=training_param['learning_rate'],
        fp16=True,
        max_grad_norm=0.3,
        logging_steps=10,
        push_to_hub=False,
        save_steps=500,
        save_total_limit=2,
        report_to=["wandb"],
    ),
)

mutiturn_trainer.train()
mutiturn_trainer.model.save_pretrained(folder_name)
```
